```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("loan_approval_dataset.csv")

data.head()
```

```python


# Example binary encoding
data[' education'] = data[' education'].map({' Graduate': 1, ' Not Graduate': 0})
data[' self_employed'] = data[' self_employed'].map({' Yes': 1, ' No': 0})
data[' loan_status'] = data[' loan_status'].map({' Approved': 1, ' Rejected': 0})

data.head()
```

```python
enc_data = data.drop('loan_id', axis=1)
enc_data.head()
```

```python
# Defining x and y
x = enc_data.drop(" loan_status", axis=1)
y = enc_data[" loan_status"]
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
# Build the model
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(8),
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

model_1.fit(x_train, y_train, epochs=100)
```

```python
# Increase number of neurons
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

model_2.fit(x_train, y_train, epochs=100)
```

```python

tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_3.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

history = model_3.fit(x_train, y_train, epochs=100)
```

Still have low accuracy and high loss values. SO let's try normalize tha data

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler

ct = make_column_transformer(
    (MinMaxScaler(), [" no_of_dependents", " income_annum", " loan_amount", " loan_term", " cibil_score", " residential_assets_value", " commercial_assets_value", " luxury_assets_value", " bank_asset_value"]), # get all values between 0 and 1
)

# Fit column transformer in train data
ct.fit(x_train)

# Transform trainiinig and test data with normalization
x_train_norm = ct.transform(x_train)
x_test_norm = ct.transform(x_test)
```

```python
x_train_norm[0]
```

```python
tf.random.set_seed(42)

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_4.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

history = model_4.fit(x_train_norm, y_train, epochs=100)
```

```python
model_4.evaluate(x_test_norm, y_test)
```

```python
y_test
```

```python
# Make predictions on the test data
predictions = model_4.predict(x_test_norm)

# Since the model outputs raw values (no sigmoid in the output layer),
# we'll apply a threshold (e.g., 0.5) to convert them to binary predictions.
# You might need to adjust this threshold based on your desired balance of precision and recall.
binary_predictions = (predictions > 0.5).astype(int)

# Map the binary predictions back to 'Approved' and 'Rejected' labels
predicted_labels = np.where(binary_predictions == 1, 'Approved', 'Rejected')

# Display the first few predictions
print("First 10 predictions (binary):")
print(binary_predictions[:10])
print("\nFirst 10 predicted labels:")
print(predicted_labels[:10])
```

```python
# Make binary_predictions 1-dimensional for comparison
binary_predictions_flat = binary_predictions.flatten()

# Now compare
y_test == binary_predictions_flat
```

```python
# Add activation functions
tf.random.set_seed(42)

model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model_5.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

history = model_5.fit(x_train_norm, y_train, epochs=100)
```

```python
model_5.evaluate(x_test_norm, y_test)
```

```python
# Try lowering the Adam() optimizer learning rate to 0.0005

tf.random.set_seed(42)

model_6 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model_6.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
                metrics=["accuracy"])

history = model_6.fit(x_train_norm, y_train, epochs=100)
```

```python
model_6.evaluate(x_test_norm, y_test)
```
